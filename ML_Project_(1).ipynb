```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from sklearn.utils import resample
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D  # Importing 3D plotting library

# Assuming 'data' is your DataFrame
# Drop non-numeric columns or encode them appropriately
numeric_data = data.select_dtypes(include=[np.number])

# Choose the number of clusters (you may need to tune this parameter)
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(data)

# Evaluate clustering performance
silhouette_avg = silhouette_score(data, cluster_labels)
calinski_harabasz_avg = calinski_harabasz_score(data, cluster_labels)

print(f"Silhouette Score: {silhouette_avg}")
print(f"Calinski-Harabasz Index: {calinski_harabasz_avg}")

# Resampling and cluster stability evaluation
num_resamples = 100
mean_similarity = []

for _ in range(num_resamples):
    # Resample the data
    resampled_data = resample(data, replace=True, random_state=_)

    # Apply clustering algorithm to resampled data
    resampled_cluster_labels = kmeans.predict(resampled_data)

    # Compute similarity for each cluster
    similarity = []
    for i in range(num_clusters):
        original_cluster_center = kmeans.cluster_centers_[i]
        resampled_cluster_centers = np.mean(resampled_data[resampled_cluster_labels == i], axis=0)

        # Calculate similarity using cosine similarity, you can choose other metrics as well
        similarity.append(1 - cdist([original_cluster_center], [resampled_cluster_centers], 'cosine')[0][0])

    # Compute mean similarity across all clusters
    mean_similarity.append(np.mean(similarity))

# Compute the mean stability across all resamples for each cluster
mean_stability = np.mean(mean_similarity)

print(f"Mean Cluster Stability: {mean_stability}")

# 2D PCA visualization
pca_2d = PCA(n_components=2)
data_2d = pca_2d.fit_transform(data)

plt.figure(figsize=(8, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.title('2D PCA Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# 3D PCA visualization
pca_3d = PCA(n_components=3)
data_3d = pca_3d.fit_transform(data)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(data_3d[:, 0], data_3d[:, 1], data_3d[:, 2], c=cluster_labels, cmap='viridis', alpha=0.5)
ax.set_title('3D PCA Visualization of Clusters')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```

# Repeat K times and compute the mean simalirity across all resampled sets for each cluster


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import KFold
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
data = df2

# Identify high cardinality columns
high_cardinality_columns = [col for col in data.columns if data[col].nunique() > 35]

# Exclude columns with high cardinality
low_cardinality_data = data.drop(columns=high_cardinality_columns)

# Standardize the data
scaler = StandardScaler()
standardized_data = scaler.fit_transform(low_cardinality_data)

# Choose the number of clusters for the original clustering
num_clusters_original = 4  # Adjust this to match the desired number of clusters

# Train the original clustering model
kmeans_original = KMeans(n_clusters=num_clusters_original, random_state=42)
cluster_labels_original = kmeans_original.fit_predict(standardized_data)

# Store original cluster centers
original_cluster_centers = kmeans_original.cluster_centers_

# Set up k-fold cross-validation for resampling
num_resamples = 100
num_splits = 4  # You can adjust the number of splits
num_repeats = 5  # Number of repeats
kf = KFold(n_splits=num_splits, shuffle=True, random_state=42)

# List to store mean similarity scores for each cluster across repeats
mean_similarity_scores = []

# Repeat K times
for repeat in range(num_repeats):
    # Perform k-fold cross-validation for resampling and clustering
    resampled_cluster_centers_list = []

    for train_index, test_index in kf.split(standardized_data):
        train_data, test_data = standardized_data[train_index], standardized_data[test_index]

        # Train the clustering model on the training data
        kmeans_resampled = KMeans(n_clusters=num_clusters_original, random_state=42)
        cluster_labels_resampled = kmeans_resampled.fit_predict(train_data)

        # Store resampled cluster centers
        resampled_cluster_centers_list.append(kmeans_resampled.cluster_centers_)

    # Compute mean similarity scores across all resampled sets for each cluster
    mean_similarity_scores_repeat = []

    for i in range(num_clusters_original):
        original_center = original_cluster_centers[i]

        similarity_scores_cluster = []
        for resampled_centers in resampled_cluster_centers_list:
            similarity = 1 - cdist([original_center], [resampled_centers[i]], 'cosine')[0]
            similarity_scores_cluster.append(similarity[0])

        mean_similarity_score = np.mean(similarity_scores_cluster)
        mean_similarity_scores_repeat.append(mean_similarity_score)

    mean_similarity_scores.append(mean_similarity_scores_repeat)

    # Print mean similarity scores for each cluster across repeats
    print(f"\nMean Similarity Scores for Each Cluster (Repeat {repeat + 1}):")
    for cluster_idx, mean_similarity in enumerate(mean_similarity_scores_repeat):
        print(f"Cluster {cluster_idx}: {mean_similarity}")

# Convert the list of mean similarity scores to a NumPy array
mean_similarity_scores = np.array(mean_similarity_scores)

# Print final mean similarity scores across all repeats
final_mean_similarity_scores = np.mean(mean_similarity_scores, axis=0)
print("\nFinal Mean Similarity Scores Across All Repeats:")
for cluster_idx, final_mean_similarity in enumerate(final_mean_similarity_scores):
    print(f"Cluster {cluster_idx}: {final_mean_similarity}")

# Plot mean similarity scores for each cluster across repeats
for i in range(num_clusters_original):
    plt.plot(range(1, num_repeats + 1), mean_similarity_scores[:, i], label=f'Cluster {i}')

plt.xlabel('Repeat')
plt.ylabel('Mean Similarity Score')
plt.title(f'Mean Similarity Scores for Each Cluster Across {num_repeats} Repeats')
plt.legend()
plt.show()
```
